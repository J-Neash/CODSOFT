```python
dataset = pd.read_csv('advertising.csv')
```

```python
print(dataset.shape)
print(dataset.head(5))
```

```python
dataset.describe()
```

```python
sns.pairplot(dataset, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', kind='scatter')
plt.suptitle("TV, Radio, and Newspaper vs. Sales")
plt.show()
```

```python
X = dataset.drop(["Sales"],axis='columns')
X
```

```python
Y = dataset.Sales
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25,random_state =0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
TV = float(input("Enter the money spent on advertising through TV: "))
Radio = float(input("Enter the money spent on advertising through Radio: "))
Newspaper = float(input("Enter the money spent on advertising through Newspaper: "))
Sales = [[TV,Radio,Newspaper]]
result = model.predict(Sales)
print(result)
```

```python
from sklearn.metrics import r2_score
r2score = r2_score(y_test, y_pred)
print("Accuracy of the Model: {0}%".format(r2score*100))
```
